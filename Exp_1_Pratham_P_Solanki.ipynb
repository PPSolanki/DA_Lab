```python
sns.heatmap(hi.corr())
```

Afterwards, we find the correlation matrix and plot it on a heatmap. 
Higher the correlation value, the similar the attributes are. The diagonal 
of such matrices will always be 1 because an attribute [link text](https://)is completely 
similar to itself. With this heatmap, we are able to see that the attribute "Ladder score in Dystopia" is same for every country. So we will drop that attribute completely as it does not give us any important information.

```python
data = hi.drop(['Ladder score in Dystopia'], axis = 1)
```

```python
sns.heatmap(data.corr())
```

```python
sns.relplot(x = 'Ladder score', y = 'Country name', data = data, height = 20, aspect = 1)
```

Next we plotted the Ladder score with respect to the countries and we saw that Finland has the highest Ladder score whilst Afghanistan has the lowest.

```python
sns.relplot(x = 'Logged GDP per capita', y = 'Ladder score', data = data)
```

When plotting the relation between Ladder score and GDP per capita, we can see that the relation is directly proportional, as one attribute increases the other does too, of course with some exceptions.

```python
sns.relplot(x = 'Healthy life expectancy', y = 'Ladder score', data = data)
```

We see a very similar kind of pattern as well whn we plot the relation between Ladder score and Health life expectancy.

```python
sns.jointplot(x = "Ladder score", y = "Logged GDP per capita", data = data, kind="kde")
```
